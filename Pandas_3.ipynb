```python
data['gender'].value_counts()
```

convert male to 0

convert female to 1

```python
def encode(x):
    if x == "Female":
        return 1
    else:
        return 0
```

#### df.apply / series.apply function

```python
data.gender = data['gender'].apply(encode) # application of apply function
data.head()
```
